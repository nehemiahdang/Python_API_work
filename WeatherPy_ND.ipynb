```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import random
from pprint import pprint
import csv

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it's being processed (with the city number and city name).


```python
# Pulling up a list of the random cities selected to double check
list(cities)
```

```python
# Got the url from the website saved it into a variable
url = "http://api.openweathermap.org/data/2.5/weather?"
#query = f'{url}appid={weather_api_key}&q={cities}'

# Created individual lists for each of the datasets I'm going to be collecting from the json file
# Created another list for cities to filter out the cities with no data
cities_found = []
cloud_data = []
country_data = []
date_data = []
humidity_data = []
lat_data = []
lng_data = []
max_temp_data = []
wind_speed_data = []

# Created a print to let me know when the loop starts and ends
print("Beginning Data Retrieval")
print("-"*30)

# Created an iteration to loop through the cities, appending all the necessary data I need
#for city in random.sample(cities, 5):
for city in cities:
    query = f'{url}appid={weather_api_key}&q={city}'
    weather_response = requests.get(query).json()
    
    try:
        cities_found.append(weather_response['name'])
        cloud_data.append(weather_response['clouds']['all'])
        country_data.append(weather_response['sys']['country'])
        date_data.append(weather_response['dt'])
        humidity_data.append(weather_response['main']['humidity'])
        lat_data.append(weather_response['coord']['lat'])
        lng_data.append(weather_response['coord']['lon'])
        max_temp_data.append(weather_response['main']['temp_max'])
        wind_speed_data.append(weather_response['wind']['speed'])
        print(f"{city} found! Inputting")
    except:
        print(f"{city} not found. Skipping")
        
print("-"*30)
print("Data Retrieval Complete")
print("-"*30)
```

```python
# Converting the max temp data from Kelvin to Fahrenheit and round up the values
fahrenheit_data = [round((x*9/5)-459.67,2) for x in max_temp_data]
fahrenheit_data
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
```

```python
# Create Northern and Southern Hemisphere DataFrames
north_df = weather_df.loc[weather_df["Lat"] >= 0, ["City","Cloudiness","Country","Date","Humidity","Lat","Lng","Max Temp", "Wind Speed"]]
south_df = weather_df.loc[weather_df["Lat"] < 0, ["City","Cloudiness","Country","Date","Humidity","Lat","Lng","Max Temp", "Wind Speed"]]
#"City","Cloudiness","Country","Date","Humidity","Lat","Lng","Max Temp", "Wind Speed"
```

####  Northern Hemisphere -  Latitude Linear vs. Max Temp Regression

```python
north_df.head()
```

```python
# Creating a stored variable with latitude and max temp
north_lat = north_df["Lat"]
north_max_temp = north_df["Max Temp"]

# Getting the linear regression the latitude as the x and max temp as the y
(slope, intercept, rvalue, pvalue, stderr) = linregress(north_lat, north_max_temp)

# Getting the regression values with the x-value as latitude
regress_values = north_lat * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Plotting the graph with their respective labels
plt.scatter(north_lat, north_max_temp)
plt.xlabel("Latitude")
plt.ylabel("Max Temp")

# Plotting the regression line onto the graph
plt.plot(north_lat, regress_values, "r-")

# Printing the line equation onto the graph and print the r-value
plt.annotate(line_eq, (7, 15), fontsize=15, color="red")
print(f"The r-squared is: {rvalue}")
plt.show()
```

```python
# There's a strong negative correlation between the latitude and temperature
# in the northern hemisphere.
```
